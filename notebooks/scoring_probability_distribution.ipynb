```python
def probscr(S):
    d = 0.03291
    T = 2880
    s = 2.0894
    return ((1/s) * ((((d*T)**(S/s))* np.exp((-d)*T)) / (gamma(S/s))))
```

```python
def grf(B,E):
    X = []
    Y = []
    while B <= E:
        X.append (B)
        Y.append (probscr(B))
        B = B + 1
        
    x = np.array(X)
    y = np.array(Y)
    plt.scatter(x, y)
    plt.show()
```

```python
#probability that many points in a row 
def Q(s):
    P = 0
    n = 1
    q = 0.348
    
    pnts = [0.087,0.7386,0.1728,0.0014]
    w1 = pnts[0]
    w2 = pnts[1]
    w3 = pnts[2]
    w4 = pnts[3]
    
    Q1 = ((1-q)*w1)
    Q2 = ((1-q)*(w2 + q*(w1**2)))
    Q3 = ((1-q)*(w3 + 2*q*w2*w1 + (q**2)*(w1**3)))
    Q4 = ((1-q)*(w4 + q*(2*w3*w1 + (w2**2)) + 3*(q**2)*w2*(w1**2) + (q**3)*(w1**4)))
    
    if s == 1:
        return Q1
    elif s == 2:
        return Q2
    elif s == 3:
        return Q3
    elif s == 4:
        return Q4
    
    else:
        for w in pnts:
            if (s-n) == 1:
                P = P + w * Q1
            elif (s-n) == 2:
                P = P +  w * Q2
            elif (s-n) == 3:
                P = P + w * Q3
            elif (s-n) == 4:
                P = P + w * Q4
            else:
                P = P + (w*(Q(s-n)))
            n = n + 1
        #print('loop')
    return (P*q)
```

```python
def Pteam(a, b, A, B, T):
    G = abs(A-B)
    r = 0
    Ia = a/(a+b)
    Ib = b/(a+b)
    if T == 1:
        r = 1
        #print ('woo')      
    elif T == 2:
        r = -1
        #print ('cool')
    elif T == 0:
        r = 0
        #print('first score ooo')
    else:
        return ('boo theres only 2 teams you goof')
    Pa = Ia - 0.152 * r - 0.0022 * G
    Pb = Ib + 0.152 * r + 0.0022 * G
    #print ('the probability that team a scores is', (Pa*100),'% \n', 'the probability that team b scores is', (Pb*100),'%')
    return (Pa,Pb)
```

```python
def PScore(A,B):
    n=0
    Sa = 0
    Sb = 0
    Ls = 0
    while n <= 95:
        S = Pteam(A,B,Sa,Sb,Ls)
        #print(S)
        if S[0] > S[1]:
            Sa = Sa + 1
            Ls = 1
        elif S[0] < S[1]:
            Sb = Sb + 1
            Ls = 2
        n = n+1
        #print(Sa,Sb)
    return(Sa,Sb)
```

```python
def PScore2(A,B):
    n=0
    Sa = 0
    Sb = 0
    Ls = 0
    while n <= 95:
        S = Pteam(A,B,Sa,Sb,Ls)
        o = random.randrange(100)
        if S[0] * 100 > o:
            Sa = Sa + pnts()
            Ls = 1
        else:
            Sb = Sb + pnts()
            Ls = 2
        n = n+1
    print ('Team A scored', (Sa),'points \n', 'Team B scored', (Sb),'points')
    return(Sa,Sb)
```

```python
def pnts():
    r = random.randrange(1000000)
    Pr = [12,230,1400,87000,172800,738600]
    Pnt = [6,5,4,1,3,2]
    n = 0
    k = 0
    while n <= r:
        n = n + Pr[k]
        k = k + 1
    return Pnt[k-1]
```
