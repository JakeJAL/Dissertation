```python
X = music_data.drop(columns = ['genre'])
Y = music_data['genre']

model = DecisionTreeClassifier()
model.fit(X.values,Y.values)

tree.export_graphviz(model, out_file = 'music recomender.dot', 
                    feature_names = ['age', 'gender'], 
                    class_names = sorted(y.unique()), 
                    label = 'all',
                    rounded = True,
                    filled = True)
```
