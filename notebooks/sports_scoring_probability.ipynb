```python
def pnts():
    r = random.randrange(1000000)
    Pr = [12,230,1400,87000,172800,738600]
    Pnt = [6,5,4,1,3,2]
    n = 0
    k = 0
    while n <= r:
        n = n + Pr[k]
        k = k + 1
    return Pnt[k-1]
```

```python
def Shot(p1,p2):
    Turn = [p1,p2]
    score = []
    for p in range(len(Turn)):
        active = Turn[p]
        prob = 0
        r = random.randrange(100)
        
        for i in range (len(Players[active])):
            prob = prob + Players[active][i]
            
            if r < (prob):
                score.append(i)
                break
                
    print (p1 , 'scored', score[0],'\n' ,p2 , 'scored', score[1])
    return score
```

```python
def PredictChecker(Predicted, Facts):
    Correct = 0
    Wrong = 0
    Match = 0
    Winner = 0
    Amount = len(Predicted)
    for i in range(len(Predicted)):
        Both = 0
        for j in range(len(Predicted[0])):
            if Predicted[i][j] == Facts[i][j]:
                Correct = Correct + 1
                Both = Both + 1
            else:
                Wrong = Wrong + 1
                
        if Facts[i][0] > Facts[i][1]:
            if Predicted[i][0] > Predicted[i][1]:
                Winner = Winner + 1
        if Facts[i][1] > Facts[i][0]:
            if Predicted[i][1] > Predicted[i][0]:
                Winner = Winner + 1
        if Facts[i][0] == Facts[i][1]:
            if Predicted[i][0] == Predicted[i][1]:
                Winner = Winner + 1
                
        if Both == 2:
            Match = Match + 1

    SucRate = Correct/(Amount * 2)
    FailRate = Wrong/(Amount * 2)
    MatchSuc = Match/Amount
    WinSuc = Winner/Amount
    
    print ('The percentage the model got a players score correct is ', round(SucRate, 2)*100, 
          '\n The percentage the model got a players score wrong is ', round(FailRate, 2)*100,
          '\n The percentage the model got a match reults correct is ', round(MatchSuc, 2)*100,
          '\n The percentage the model got the winner correct is ', round(WinSuc, 2)*100)
    
    return (SucRate, FailRate)
```

```python
def fight3(a,b,c,d) :
    ta = [a,b]
    tb = [c,d]
    ah = a[0] - (c[1] + d[1])/len(tb)
    bh = b[0] - (c[1] + d[1])/len(tb)
    ch = c[0] - (a[1] + b[1])/len(ta)
    dh = d[0] - (a[1] + b[1])/len(ta)
    return (ah,bh,ch,dh)
```
