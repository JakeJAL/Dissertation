```python
def kill(k):
    s = (k/600) * 10000
    p = s
    t = 0
    kill = 0
    while t <= 600:
        r = random.randrange(k * 10000)
        if r <= p:
            kill = kill + 1
            print(kill,t,p)
            p = s
            t = t + 1
        else:
            #print(p)
            p = p + s
            t = t + 1
    return (kill)
```
