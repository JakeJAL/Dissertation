```python
Fears = {
    'Dave': 5,
    'Sue' : -3, 
    'Alan' : -2,
    'Blaz' : 10,
    'Brian' : -10,
    'Grog' : 0}
```

```python
def Match1(p1,p2,m):
    Turn = [p1,p2]
    score = [0,0]
    Players1 = copy.deepcopy(Players)
    
    for i in range(2):
        a = Fears[Turn[i]] * 0.7
        b = Fears[Turn[i]] * 0.3
        #print(Fears[Turn[i]],i)
        
        Players1 [Turn[i-1]][0] = Players1 [Turn[i-1]][0] + a + b/2
        Players1 [Turn[i-1]][1] = Players1 [Turn[i-1]][1] - a + b/2
        Players1 [Turn[i-1]][2] = Players1 [Turn[i-1]][2] - b
    
        #print(Turn[i-1],a,b,Players1[Turn[i]])
    
    for s in range (m):
        
        for p in range(len(Turn)):
            active = Turn[p]
            prob = 0
            r = random.randrange(10000)/100

            for i in range (len(Players1[active])):
                prob = prob + Players1[active][i]

                if r < (prob):
                    score[p] = score[p] + i
                    break

    return (score)
```
