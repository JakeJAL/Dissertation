```python
def bipart(M,N):
    mtrx = []
    n = 0

    while n < N+M:
        t = 0
        adj = []
        if n < M:
            while t < N+M:
                if t < M:
                    adj.append (0)
                else:
                    adj.append (1)
                t = t+1 
        else:
            while t < N+M:
                if t < M:
                    adj.append (1)
                else:
                    adj.append (0)
                t = t+1 
        mtrx.append (adj)
        n = n+1 
    return (mtrx)
```

```python
bipart(4,2)
```

```python
def wheel(N):
    mtrx = []
    n = 0
    while n < N+1:
        t = 0
        adj = []
        if n == 0:
            while t < N+1:
                if t == 0:
                    adj.append (0)
                else:
                    adj.append (1)
                t = t+1 
        else:
            while t < N+1:
                if t == 0:
                    adj.append (1)
                elif (n == 1 and t == N):
                    adj.append (1)
                elif (n == N and t == 1):
                    adj.append (1)
                elif t == n+1 or t == n-1 :
                    adj.append (1)
                else:
                    adj.append (0)
                t = t+1 
        mtrx.append (adj)
        n = n+1 
    return (mtrx)
```

```python
wheel (4)
```

```python
def star(N):
    mtrx = []
    n = 0

    while n < N+1:
        t = 0
        adj = []
        if n == 0:
            while t < N+1:
                if t == 0:
                    adj.append (0)
                else:
                    adj.append (1)
                t = t+1 
        else:
            while t < N+1:
                if t == 0:
                    adj.append (1)
                else:
                    adj.append (0)
                t = t+1 
        mtrx.append (adj)
        n = n+1 
    return (mtrx)
```

```python
star(4)
```

```python
def path(N):
    mtrx = []
    n = 0
    while n < N:
        t = 0
        adj = []
        while t < N:
            if t == n+1 or t == n-1 :
                adj.append (1)
            else:
                adj.append (0)
            t = t+1 
        mtrx.append (adj)
        n = n+1 
    return (mtrx)
```

```python
path(4)
```

```python
def cycle(N):
    mtrx = []
    n = 0
    while n < N:
        t = 0
        adj = []
        while t < N:
            if (n == 0 and t == N-1):
                adj.append (1)
            elif (n == N-1 and t == 0):
                adj.append (1)
            elif t == n+1 or t == n-1 :
                adj.append (1)
            else:
                adj.append (0)
            t = t+1 
        mtrx.append (adj)
        n = n+1 
    return (mtrx)
```

```python
cycle(4)
```

```python
def complete(N):
    mtrx = []
    n = 0
    while n < N:
        t = 0
        adj = []
        while t < N:
            if t == n:
                adj.append (0)
            else:
                adj.append (1)
            t = t+1 
        mtrx.append (adj)
        n = n+1 
    return (mtrx)
```
